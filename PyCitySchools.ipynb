```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
for i in prefixes_suffixes:
    students_df["student_name"] = students_df["student_name"].str.replace(i,"")
student_list = students_df.student_name.tolist()
```

```python
incorrect_students = []
for i in student_list:
    if len(i.split()) > 2:
        incorrect_students.append(i)
incorrect_students
```

```python
school_students_df = students_df.merge(school_df, how="inner", on="school_name")
school_students_df
```

```python
students = students_df["Student ID"].count()
students
```

```python
schools = len(set(school_df["school_name"]))
schools
# len(school_df["school_name"].unique())
```

```python
budget = school_df["budget"].sum()
budget
```

```python
avg_reading = students_df.reading_score.mean()
avg_reading
```

```python
avg_math = students_df.math_score.mean()
avg_math
```

```python
reading_pass_df = students_df[students_df.reading_score >= 70]
reading_pass = reading_pass_df["Student ID"].count()
reading_pass_percent = reading_pass/39170 * 100
reading_pass_percent
```

```python
math_pass_df = students_df[students_df.math_score >= 70]
math_pass = math_pass_df["Student ID"].count()
math_pass_percent = math_pass/39170 * 100
math_pass_percent
```

```python
both_pass_df = reading_pass_df.merge(math_pass_df, how="inner", on="Student ID")
both_pass = both_pass_df["Student ID"].count()
both_pass_percent = both_pass/39170 * 100
both_pass_percent
```

```python
district_summary_dict = {"Total Schools": schools,
                         "Total Students": students,
                         "Total Budget": budget,
                         "Average Reading Score": avg_reading,
                         "Average Math Score": avg_math,
                         "% Passing Reading": reading_pass_percent, 
                         "% Passing Math": math_pass_percent, 
                         "% Overall Passing": both_pass_percent
                        }
district_summary_dict
```

```python
district_summary_df = pd.DataFrame([district_summary_dict])
district_summary_df
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}%".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}%".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}%".format)
district_summary_df
```

```python
new_column_order = ["Total Students","Total Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]
district_summary_df = district_summary_df[new_column_order]
district_summary_df
```
