```python
# Aggregations
school_summary_df = school_df[["school_name","type","budget"]].set_index("school_name")
students = school_students_df.groupby(["school_name"]).count()["Student ID"]
math_score = school_students_df.groupby(["school_name"]).mean()["math_score"]
read_score = school_students_df.groupby(["school_name"]).mean()["reading_score"]
math_pass = school_students_df[school_students_df["math_score"]>=70].groupby(["school_name"]).count()["math_score"]
read_pass = school_students_df[school_students_df["reading_score"]>=70].groupby(["school_name"]).count()["reading_score"]
both_pass = school_students_df.loc[(school_students_df["reading_score"] >=70) & (school_students_df["math_score"]>=70),:].groupby(["school_name"]).count()["math_score"]

# Merge Series into DataFrame
school_summary_df["Total Students"] = students
school_summary_df["Avg Math Score"] = math_score
school_summary_df["Avg Reading Score"] = read_score
school_summary_df["Math Pass"] = math_pass
school_summary_df["Reading Pass"] = read_pass
school_summary_df["Overall Pass"] = both_pass

#Rename index & columns
school_summary_df.index.names = ['School Name']
school_summary_df = school_summary_df.rename(columns={"type":"School Type", "budget":"Total School Budget"})

# Calculations
school_summary_df["Per Student Budget"] = school_summary_df["Total School Budget"]/school_summary_df["Total Students"]
school_summary_df["% Passing Math"] = school_summary_df["Math Pass"]/school_summary_df["Total Students"] * 100
school_summary_df["% Passing Reading"] = school_summary_df["Reading Pass"]/school_summary_df["Total Students"] * 100
school_summary_df["% Overall Passing"] = school_summary_df["Overall Pass"]/school_summary_df["Total Students"] * 100
school_summary_df = school_summary_df.drop(columns=["Math Pass","Reading Pass","Overall Pass"])

# Reorder & Drop columns
school_summary_df = school_summary_df[["School Type","Total Students","Total School Budget","Per Student Budget","Avg Math Score",
                                      "Avg Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
school_summary_df = school_summary_df.sort_values(["School Name"])

# Formatting
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df.head()
```
