```python
student_list = students_df.student_name.tolist()
```

```python
incorrect_students = []
for i in student_list:
    if len(i.split()) > 2:
        incorrect_students.append(i)
len(incorrect_students)
```

```python
prefixes = []
for i in incorrect_students:
    if len(i.split()[0]) <= 4:
        prefixes.append(i.split()[0])
set(prefixes)
```

```python
suffixes = []
for i in incorrect_students:
    if len(i.split()[-1]) <= 3:
        suffixes.append(i.split()[-1])
set(suffixes)
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
for i in prefixes_suffixes:
    students_df["student_name"] = students_df["student_name"].str.replace(i,"")
student_list = students_df.student_name.tolist()
incorrect_students = []
for i in student_list:
    if len(i.split()) > 2:
        incorrect_students.append(i)
incorrect_students
```

```python
school_students_df = school_df.merge(students_df, how="inner", on="school_name")
school_students_df
```
